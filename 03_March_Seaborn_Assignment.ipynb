Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer:-
Five plots in Seaborn along with their usage is as following :

1.Box Plot : box plot (or box-and-whisker plot) displays the distribution of a dataset by showing the median, quartiles, and potential outliers. It is useful for comparing distributions between different groups.

2.Heatmap : A heatmap is a graphical representation of data that uses color-coding to represent the values of a two-dimensional dataset. It is used to identify patterns in the data, such as correlations or clusters.

3.Pair Plot : A pair plot is a graphical representation of data that shows pairwise relationships between different variables in a dataset. It is used to identify correlations between variables and to explore the relationships between them.

4.Violin Plot : A violin plot combines aspects of a box plot and a density plot. It shows the distribution of the data across different categories, displaying both the probability density and the summary statistics. It is useful for comparing multiple categories.

5.Joint Plot : A joint plot is a graphical representation of data that shows the relationship between two variables using a scatter plot and a histogram. It is used to explore the relationship between two variables and to identify correlations between them.

These are just a few examples of the types of plots that can be created using Seaborn. Seaborn offers many other visualizations that can be used to explore and analyze different types of data.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

Answer:-Let's start by loading the "fmri" dataset using the load_dataset function from Seaborn and then plot a line plot using timepoint as the x-axis and signal as the y-axis for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Add title and labels
plt.title('FMRI Signal Over Time')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Display the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

Answer:-Start by loading the "titanic" dataset using the load_dataset function from Seaborn and then plot two box plots: one for pclass vs age and another for pclass vs fare.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Box plot for pclass vs age
sns.boxplot(x="pclass", y="age", data=titanic, ax=axs[0])
axs[0].set_title('Age Distribution by Class')
axs[0].set_xlabel('Passenger Class')
axs[0].set_ylabel('Age')

# Box plot for pclass vs fare
sns.boxplot(x="pclass", y="fare", data=titanic, ax=axs[1])
axs[1].set_title('Fare Distribution by Class')
axs[1].set_xlabel('Passenger Class')
axs[1].set_ylabel('Fare')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

Answer:-Let's use the "diamonds" dataset from Seaborn to plot a histogram for the 'price' column and use the hue parameter for the 'cut' column



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram for the 'price' column with hue based on 'cut'
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack", palette="viridis")

# Add title and labels
plt.title('Price Distribution by Diamond Cut')
plt.xlabel('Price')
plt.ylabel('Count')

# Display the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

Answer:-Let's use the "iris" dataset from Seaborn to create a pair plot, using the hue parameter to differentiate between the species of the iris flowers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a pair plot
sns.pairplot(iris, hue="species", palette="Set2")

# Add title
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)

# Display the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

Answer:-Let's use the "flights" dataset from Seaborn to create a heatmap

```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights.head()
```

Pivot the dataset to create a matrix with month on the x-axis, year on the y-axis, and passengers as values

```python
flights_matrix = flights.pivot(index="year", columns="month", values="passengers")
flights_matrix
```

Creating heatmap using Seaborn

```python
import matplotlib.pyplot as plt
sns.heatmap(flights_matrix,
            cmap="YlGnBu",
            annot=True,
            fmt="d",
            linewidths=.5,
            linecolor="black")
plt.title("Passengers per Month")
plt.show()
```
